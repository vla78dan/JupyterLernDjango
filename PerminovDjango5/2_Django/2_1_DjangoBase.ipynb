Что такое Django?
Django — это инструмент для быстрой разработки сайтов и приложений на Python. Он позволяет создавать готовые веб-сайты и приложения быстрее и проще, чем писать код с нуля. В нём есть всё необходимое:

- встроенный веб-сервер для разработки (для продакшена используется сторонний веб-сервер, например Gunicorn);
- механизмы авторизации пользователей;
- подключение и работа с базами данных;
- шаблоны страниц и простых веб-интерфейсов;
- встроенные инструменты для работы с веб-страницами;
- система кеширования для снижения нагрузки на сайт;
- мультиязычность для сайтов с переключением языков.
С помощью Django можно быстро собрать работающее веб-приложение со сложной логикой из готовых элементов.

Встроенная панель управления
Одно из главных преимуществ Django — встроенная панель управления сайтом. Разработчику не нужно создавать свою админку — она доступна сразу после запуска сайта.
В панели управления можно:

- настраивать структуру сайта;
- управлять пользователями;
- работать с фильтрами и выгружать данные;
- вносить изменения в код и видеть их на сайте;
- взаимодействовать с базой данных сайта;
- просматривать статистику.
Панель управления написана на Python, поэтому её можно оптимизировать или переделать под задачи проекта. Это одно из главных достоинств Django.

Новые функции в Django 5
В Django 5.0 упрощен вывод полей формы в шаблонах - добавлены группы полей as_field_group.

В Django 5.0 добавлены значения по умолчанию - db_default, для полей моделей, вычисляемые на стороне базы данных.

В Django 5.0 добавлены генерируемые поля для моделей - GeneratedField, на основе значений других полей, вычисляемые на стороне базы данных.

В Django 5.0 добавлены дополнительные возможности для объявления параметров выбора(choices) в полях моделей и в полях, типа ChoiceField, форм.

Django 5.0 перестал поддерживать Python 3.8 и 3.9. На данный момент поддерживает только Python 3.10, 3.11 и 3.12.
