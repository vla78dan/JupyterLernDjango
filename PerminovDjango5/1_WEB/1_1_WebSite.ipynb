Одним из ключевых протоколов, обеспечивающих функционирование Всемирной паутины, является HTTP (Hypertext Transfer Protocol)

HTTP является протоколом без сохранения состояния (stateless protocol)

URL (Uniform Resource Locator): Адрес Ресурса
URL, или Uniform Resource Locator, — это, по сути, адрес ресурса в интернете. Он указывает, где именно находится запрашиваемый файл, изображение, страница или другой веб-ресурс. Представьте URL как почтовый адрес для данных в цифровом мире.

Пример URL:

https://www.example.com/blog/article123

Давайте разберем его на части:

Протокол (https://): Это начальная часть URL, которая указывает, какой протокол используется для передачи данных. https:// (Hypertext Transfer Protocol Secure) обеспечивает безопасную передачу данных, шифруя их. Другой распространенный протокол - http://, но он не обеспечивает шифрования.
Домен (www.example.com): Это имя сайта, которое переводит понятный для человека текст в IP-адрес сервера. Доменное имя указывает на конкретный веб-сайт или ресурс в интернете. www — это часто используемый префикс, указывающий на веб-сервер.
Путь (/blog/article123): Это часть URL, которая определяет конкретное местоположение ресурса на сервере. Путь указывает на папку blog и конкретный файл article123 (или же динамический ресурс).
URL необходим для того, чтобы браузер понимал, какой именно ресурс вы хотите получить, и как к нему добраться.

HTTP-запрос (HTTP Request):
HTTP-запрос — это сообщение, которое отправляет клиент (например, ваш браузер) на сервер. Он сообщает серверу, какой ресурс запрашивается и какое действие необходимо выполнить. По сути, это “запрос” клиента к серверу на предоставление определенного ресурса.

HTTP-запрос состоит из нескольких ключевых частей:

Метод (Method): Это действие, которое клиент хочет выполнить с ресурсом. Основные HTTP-методы:
GET: Запрос на получение ресурса (например, HTML-страницы, изображения). Это самый часто используемый метод.
POST: Отправка данных на сервер (например, при отправке формы).
PUT: Обновление существующего ресурса.
DELETE: Удаление ресурса.
Заголовки (Headers): Метаданные запроса, предоставляющие дополнительную информацию о запросе. Заголовки могут содержать информацию о типе содержимого, языке, используемом браузере, авторизационные данные и т.д.
Тело (Body): Необязательная часть запроса, которая может содержать данные, которые нужно отправить на сервер (например, данные формы в методе POST).
Пример HTTP-запроса (упрощенно):

GET /blog/article123 HTTP/1.1
Host: www.example.com
User-Agent: Mozilla/5.0 ...

В этом примере браузер (клиент) запрашивает ресурс /blog/article123 с использованием метода GET. Заголовок Host указывает на домен, а User-Agentпредоставляет информацию о браузере.

HTTP-ответ (HTTP Response): Ответ Сервера
HTTP-ответ — это сообщение, которое сервер отправляет клиенту в ответ на его запрос. Он содержит запрошенный ресурс или сообщение об ошибке, если что-то пошло не так. HTTP-ответ можно представить как “ответ” от сервера на “запрос” клиента.

HTTP-ответ состоит из следующих ключевых частей:

Код состояния (Status Code): Это числовой код, который сообщает клиенту, как был обработан запрос. Некоторые распространенные коды состояния:
200 OK: Запрос выполнен успешно, и ресурс предоставлен.
404 Not Found: Запрашиваемый ресурс не найден.
500 Internal Server Error: Ошибка на стороне сервера.
301 Moved Permanently: Ресурс перемещен навсегда.
Заголовки (Headers): Метаданные ответа, которые предоставляют дополнительную информацию о ответе (тип контента, дата, сервер и т.д.).
Тело (Body): Содержит сам запрошенный ресурс (например, HTML-код веб-страницы) или сообщение об ошибке.
Пример HTTP-ответа (упрощенно):

HTTP/1.1 200 OK
Content-Type: text/html; charset=utf-8
Date: Tue, 14 May 2024 10:00:00 GMT

<!DOCTYPE html>
<html>
  ...  (HTML код страницы) ...
</html>

В этом примере сервер отвечает кодом состояния 200 OK, указывая на успешное выполнение запроса. Заголовок Content-Type указывает, что в теле ответа находится HTML-код.

Структура HTTP-запроса:

HTTP-запрос состоит из трех основных частей:

Метод запроса (HTTP Method):

Это самая важная часть запроса, поскольку она определяет, какое действие клиент хочет выполнить на сервере. Метод указывает на тип операции, которую сервер должен предпринять с запрошенным ресурсом.

Давайте рассмотрим наиболее распространенные методы:

GET:

Используется для запроса данных с сервера, обычно для получения HTML-страниц, изображений, скриптов и т.д.
Данные передаются в URL (через строку запроса).
Это безопасный метод (не изменяет данные на сервере).
Пример: Когда вы вводите адрес веб-сайта в браузере, браузер отправляет GETзапрос на сервер, что бы получить HTML код страницы.
POST:

Используется для отправки данных на сервер для обработки.
Данные обычно передаются в теле запроса, что более безопасно и подходит для больших объемов информации.
Часто используется для отправки данных из HTML-форм (регистрация, авторизация, отправка комментариев и т.д.).
Пример: Когда вы заполняете форму обратной связи и нажимаете кнопку “Отправить”, браузер отправляет POST запрос на сервер с введенными вами данными.
PUT:

Используется для обновления существующего ресурса на сервере.
Данные, которые отправляются с PUT запросом, должны содержать полное представление ресурса, включая все его поля.
Пример: Когда вы редактируете данные профиля в приложении и сохраняете изменения, приложение может отправить PUT запрос на сервер, чтобы обновить ваши данные.
DELETE:

Используется для удаления ресурса с сервера.
Обычно не передаёт данных в теле запроса, так как идентификатор ресурса, который нужно удалить, указывается в URL.
Пример: Когда вы удаляете пост в блоге, приложение может отправить DELETEзапрос на сервер для удаления этого поста.
Другие методы:

PATCH: используется для частичного обновления ресурса.
HEAD: используется для запроса заголовков ресурса без тела ответа.
OPTIONS: используется для получения списка доступных методов для ресурса.
Менее распространены, но используются в различных API.

Заголовки запроса (Request Headers):

Заголовки запроса представляют собой пары “ключ-значение”, которые содержат дополнительную информацию о запросе. Они предоставляют контекст серверу и позволяют ему обрабатывать запрос соответствующим образом.

Некоторые важные заголовки:

Host: Указывает имя хоста сервера, к которому обращается клиент. Это необходимо, когда несколько сайтов хостятся на одном IP-адресе.
Пример: Host: www.example.com
User-Agent: Содержит информацию о браузере и операционной системе клиента. Это позволяет серверу предоставлять различный контент в зависимости от типа клиента.
Пример: User-Agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36
Accept: Указывает типы контента, которые клиент может принять. Это помогает серверу правильно форматировать ответ.
Пример: Accept: text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8
Существуют и другие заголовки, которые могут включать в себя тип контента, язык, аутентификацию и другие данные.

Тело запроса (Request Body):

Тело запроса содержит данные, которые клиент отправляет на сервер. Это может быть, например, текст, JSON, XML, данные из формы или файлы.
Используется преимущественно с методами POST и PUT, реже с другими методами.
Пример: При отправке формы с логином и паролем, данные из формы будут переданы в теле запроса.
Тело запроса не используется при GET и DELETE методах.
